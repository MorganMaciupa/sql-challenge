```python
#Dependencies and setup
from sqlalchemy import create_engine
import psycopg2
import matplotlib.pyplot as plt
import pandas as pd
from password import password

engine = create_engine('postgresql://localhost:5432/employee_db')

# Connect to postgres DB
conn = psycopg2.connect(
    host="localhost",
    database="employee_db",
    user="postgres",
    password= password)
```

```python
# Open a cursor to perform database operations
cur = conn.cursor()

# Execute a query
cur.execute("SELECT * FROM salaries")

# Retrieve query results
salaries_data = cur.fetchall()
salaries_data
```

```python
df = pd.DataFrame(salaries_data, columns=['emp_no', 'salary'])
df

plt.hist(df['salary']);
plt.ylabel("Frequency")
plt.xlabel("Salary ($)");
```

```python
# Open a cursor to perform database operations
cur2 = conn.cursor()

# Execute a query
cur2.execute("select * from salary_by_title")

# Retrieve query results
salary_title = cur2.fetchall()
salary_title
```

```python
df2 = pd.DataFrame(salary_title, columns=['title', 'salary'])

df2["salary"] = pd.to_numeric(df2["salary"], errors='coerce')

plt.barh(df2['title'], df2['salary'])
plt.xlabel("Average Salary ($)")
plt.ylabel("Title")
plt.title("Average Salary by Title");


```
